# Import Libraries

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

# Select() :-

```python
df.select(col("Item_Fat_Content")).show()
```

# Using Alias() :-

```python
df.select(col('Item_Identifier').alias('Item_Id')).show()
```

```python
df
```

# Filter() :-

```python
df.filter(col("Item_Fat_Content")=='Regular').show()
```

```python
df.select(col("Item_Fat_Content"),col("Item_Identifier").alias('Item_Id')).show()
```

```python
df.filter(col('Item_Identifier')=='FDA15').show()
```

```python
df.filter(col('Item_Identifier')=='FDA15').display()
```

# withColumnRenamed() :-

```python
df=df.withColumnRenamed('Item_Identifier','Item_Id')
```

```python
df
```

# Add columns

```python
df=df.withColumn('New_col_for_test',lit('None'))
```

```python
df.select('New_col_for_test').show()
```

```python
df=df.withColumn('New_Product_MRP',col("Item_MRP")*col("Item_Weight"))
```

```python
df
```

```python
df.select('New_Product_MRP').show()
```

```python
df.withColumn('Item_Fat_Content',regexp_replace(col('Item_Fat_Content'),'Regular','reg'))\
    .withColumn('Item_Fat_Content',regexp_replace(col('Item_Fat_Content'),'Low Fat','lf')).display()
```

# Casting() :-

```python
df
```
