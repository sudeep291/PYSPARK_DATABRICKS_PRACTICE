```python
df.groupBy('Item_Type').agg(sum("Item_Weight").alias('Sum_item_Weight'),max('Item_Weight').alias('Max_item_Weight'),min('Item_Weight').alias('Min_item_Weight'),avg("Item_Weight").alias('Avg_item_Weight')).display()
```

# Aggregations

```python
aggregations=df.groupBy('Item_Type').agg(sum("Item_Weight").alias('Sum_item_Weight'),max('Item_Weight').alias('Max_item_Weight'),min('Item_Weight').alias('Min_item_Weight'),avg("Item_Weight").alias('Avg_item_Weight'))
```

```python
aggregations.display()
```

```python
aggregations.limit(5).display()
```

```python
aggregations.sort("Avg_item_Weight",ascending=False).display()
```
