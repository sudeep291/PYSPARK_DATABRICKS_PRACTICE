```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df=spark.table('big_mart_sales')
```

```python
df.limit(5).display()
```

```python
df=df.withColumn('new_col',split('Outlet_Type',' '))
```

```python
df.limit(5).display()
```

```python
df.select(explode("new_col")).display()
```

```python
mine=df.select(col('Outlet_Location_Type').alias('c1'),explode("new_col").alias('col'))
```

```python
mine.limit(4).display()
```

# Collect_List()

```python
mine.groupBy('c1').agg(collect_list('col')).display()
```

```python
mine1=mine.groupBy('c1').agg(collect_list('col'))
```

```python
mine1.limit(2).display()
```

```python
mine1.select('c1',explode('collect_list(col)')).display()
```

```python
df.limit(6).display()
```

# Pivot()

```python
df.groupBy("Item_Type").pivot('Outlet_Size').agg(avg("Item_MRP")).limit(4).display()
```

```python
df.limit(5).display()
```

```python
df.groupBy('Item_Type').pivot('Item_Fat_Content').agg(sum('Item_MRP')).limit(10).display()
```
