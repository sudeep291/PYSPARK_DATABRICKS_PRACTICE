# Joins


```python
employees_data = [
    (1, "Arun", 10, 50000),
    (2, "Balu", 20, 60000),
    (3, "Charan", 30, 70000),
    (4, "Divya", 40, 80000),
    (5, "Esha", 50, 45000),
    (6, "Farhan", 20, 55000)
]

departments_data = [
    (10, "Data", "Hyderabad"),
    (20, "Engineering", "Bangalore"),
    (30, "Analytics", "Chennai"),
    (60, "HR", "Delhi")
]


df1 = spark.createDataFrame(
    employees_data,
    ["emp_id","emp_name","dept_id","salary"]
)

df2 = spark.createDataFrame(
    departments_data,
    ["dept_id","dept_name","location"]
)
```

```python
df1.limit(2).show()
```

```python
df2.limit(2).show()
```

# **Inner Join**

```python
df1.join(df2,df1.dept_id==df2.dept_id,'inner').show()
```

# Left Join

```python
df1.join(df2,df1.dept_id==df2.dept_id,'left').show()
```

# Right Join

```python
df1.join(df2,df1.dept_id==df2.dept_id,'right').show()
```

# Full Join

```python
df1.join(df2,df1.dept_id==df2.dept_id,'full').show()
```

# Outer Join

```python
df1.join(df2,df1.dept_id==df2.dept_id,'left_semi').show()
```

```python
df1.show()
```

```python
df1.join(df2,df1.dept_id==df2.dept_id,'left_anti').show()
```

```python
df1.crossJoin(df2).show()
```

```python
df1.join(df2,df1.dept_id==df2.dept_id,'leftouter').show()
```
