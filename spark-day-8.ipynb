#Window functions


```python
from pyspark.sql.window import Window
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df=spark.table('big_mart_sales')
```

```python
df.limit(5).display()
```

```python
df=df.withColumn('Rank',rank().over(window=Window.partitionBy('Outlet_Establishment_Year').orderBy(col('Outlet_Identifier').desc())))
```

```python
df.filter(col('Rank')==2).display()
```

```python
df.select('Rank').limit(30).display()
```

```python
df.display()
```

# Row_Number()

```python
df.withColumn('Row_Number',row_number().over(Window.orderBy(col('Item_Outlet_Sales').desc()))).display()
```

```python
df.withColumn('Row_Number',row_number().over(Window.orderBy(col('Rank')))).limit(4).display()
```

```python
df=df.withColumn('Row_Number',row_number().over(Window.orderBy(col('Rank'))))
```

```python
df=df.withColumn('Dense_Rank',dense_rank().over(window=Window.partitionBy('Outlet_Establishment_Year').orderBy(col('Outlet_Identifier').desc())))
```

```python
df.filter(col("Dense_Rank")==2).display()
```

```python
df.limit(4).display()
```

# Cumulative_Sum or moving or running sum

```python
df.withColumn('Cum_Sum',sum('Item_MRP').over(Window.partitionBy('Outlet_Type').orderBy(col('Outlet_Identifier').desc()).rowsBetween(Window.unboundedPreceding,Window.currentRow))).display()
```

# User_Defined_Functions(UDF)

```python
def my_fun(x):
    return x*x
```

```python
my_udf=udf(my_fun)
```

```python
df.limit(4).display()
```

```python
df.withColumn('Square',my_udf('Item_MRP')).display()
```
