# Writing modes

###  1.append-adds the file if the data already occurs or not

```python
df.write.format('csv')\
    .option('header',True)\
        .mode('append')\
            .save('/Volumes/workspace/default/my_files/data.csv')
```

## 2.overwrite

```python
df.write.format('csv')\
    .option('header',True)\
        .mode('overwrite')\
            .save('/Volumes/workspace/default/my_files/data.csv')
```

## 3.Error

```python
df.write.format('csv')\
    .option('header',True)\
        .mode('error')\
            .save('/Volumes/workspace/default/my_files/data.csv')
```

## 4.Ignore

```python
df.write.format('csv')\
    .option('header',True)\
        .mode('ignore')\
            .save('/Volumes/workspace/default/my_files/data.csv')
```
